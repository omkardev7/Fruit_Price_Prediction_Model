```python
import pandas as pd
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split
import joblib
```

```python
df = pd.read_csv('./dataset/Fruit_Dataset_2.csv')
df.sample(7)
```

```python
# Group the data by Centername and months and take the mean of the fruit prices for each group
df_grouped = df.groupby(['Centername', 'Month']).mean().reset_index()
df_grouped.tail(13)
```

```python
X = df_grouped.iloc[:, [0, 1]]
y = df_grouped.drop(['Centername', 'Month'], axis=1)
```

```python
# One-hot encode the categorical variables
X_encoded = pd.get_dummies(X, columns=['Centername', 'Month'])
```

```python
X
```

```python
X_encoded
```

```python
# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X_encoded, y, test_size=0.2, random_state=42)
```

```python
y_train
```

```python
# Create a random forest regressor model
rf = RandomForestRegressor(n_estimators=200, random_state=42)
rf.fit(X_train, y_train)

y_pred = rf.predict(X_test)
y_pred

# Evaluate the performance of the model on the test data using R-squared
from sklearn.metrics import r2_score
r2 = r2_score(y_test, y_pred)
print("R-squared:", r2)
```

```python
from sklearn.neighbors import KNeighborsRegressor
from sklearn.metrics import mean_squared_error
knn = KNeighborsRegressor(n_neighbors=5)

# Fit the model to the training data
knn.fit(X_train, y_train)

# Make predictions on the testing data
y_pred2 = knn.predict(X_test)
 

# Evaluate the model performance using Root Mean Squared Error (RMSE)
rmse = mean_squared_error(y_test, y_pred2, squared=False)
print("RMSE:", rmse)

r2 = r2_score(y_test, y_pred2)
print("R-squared:", r2)
```

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()

# Fit the model to the training data
lr.fit(X_train, y_train)

# Make predictions on the testing data
y_pred3 = lr.predict(X_test)

# Evaluate the model performance using Root Mean Squared Error (RMSE)
rmse = mean_squared_error(y_test, y_pred3, squared=False)
print("RMSE:", rmse)

r2 = r2_score(y_test, y_pred3)
print("R-squared:", r2)
```
